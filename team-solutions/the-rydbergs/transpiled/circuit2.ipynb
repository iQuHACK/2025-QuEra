```python
from qiskit import QuantumCircuit, transpile
from math import pi

circ = QuantumCircuit(3);
circ.h(2)
circ.crz(pi / 2, 1, 2)
circ.crz(pi / 4, 0, 2)
circ.h(1)
circ.crz(pi / 2, 0, 1)
circ.h(0)

circ.draw('mpl')
```

```python
basis_gates = ['r', 'rz', 'cz', 'grz', 'gr']

circ_tr = transpile(circ, basis_gates=basis_gates, optimization_level=3)
circ_tr.draw('mpl')
```

```python
from qiskit import qasm3

print(qasm3.dumps(circ_tr))
```
