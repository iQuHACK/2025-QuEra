```python
from qiskit import QuantumCircuit, transpile
from math import pi

circ = QuantumCircuit(7);

circ.h(1)
circ.h(2)
circ.h(3)
circ.cx(6, 5)
circ.cx(1, 0)
circ.cx(2, 4)
circ.cx(2, 0)
circ.cx(3, 5)
circ.cx(1, 5)
circ.cx(6, 4)
circ.cx(2, 6)
circ.cx(3, 4)
circ.cx(3, 0)
circ.cx(1, 6)

circ.draw('mpl')
```

```python
basis_gates = ['r', 'rz', 'cz']

circ_tr = transpile(circ, basis_gates=basis_gates, optimization_level=3)
circ_tr.draw('mpl')
```

```python
from qiskit import qasm3

print(qasm3.dumps(circ_tr))
```
