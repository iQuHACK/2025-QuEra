```python
from qiskit import QuantumCircuit, transpile
from math import pi

circ = QuantumCircuit(9);

circ.cx(0, 3)
circ.cx(0, 6)
circ.h(0)
circ.h(3)
circ.h(6)
circ.cx(0, 1)
circ.cx(3, 4)
circ.cx(6, 7)
circ.cx(0, 2)
circ.cx(3, 5)
circ.cx(6, 8)

circ.draw('mpl')
```

```python
basis_gates = ['r', 'rz', 'cz']

circ_tr = transpile(circ, basis_gates=basis_gates, optimization_level=3)
circ_tr.draw('mpl')
```

```python
from qiskit import qasm3

print(qasm3.dumps(circ_tr))
```
